## 002 Rock-Paper-Scissors
2023.9.29

```python
'''
勝利条件
0, 1
1, 2
2, 0
'''
import random

you = 0
cpu = 0
options = ['グー', 'チョキ', 'パー']

print('じゃんけんをしましょう。')
while you == cpu :
  cpu = int(random.randint(0,2))
  while 0 <= you and you <= 2 :
    you = int(input('何を出すか教えてください。\n 0:グー\n 1:チョキ\n 2:パー\n 9:終了\n'))
    if you == 9 :
      print ('ゲームを終了します')
      break
    elif you < 0 or 2 < you :
      print('予期せぬ入力がありました。入力をやり直してください。')
      you = 0
    else :
      print('  あなた > ' + options[you])
      print('  CPU > ' + options[cpu])

      if you == cpu :
        print('引き分けです')
        continue

      if you >= cpu % 2 :
        print('あなたの負けです')
      else :
        print('あなたの勝ちです')

      ret = int(input('もう一度遊びますか? \n 0:No \n 1:Yes\n'))
      if ret == 0 :
        print('ゲームを終了します')
        break
```
